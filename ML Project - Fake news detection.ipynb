```python
# Reference - https://www.datasnips.com/58/remove-stop-words-from-text-in-dataframe-column/

# Remove stop words
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
df['title'] = df['title'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df['text'] = df['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df.head(10)
```

### Feature Extraction - Count Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
```

```python
# Split data
x_train, x_test, y_train, y_test = train_test_split(df['text'], df['target'], test_size=0.2, random_state=0)
```

```python
x_train
```

```python
tfidf = TfidfVectorizer()
x_train_tfidf = tfidf.fit_transform(x_train) # Convert training data to vectors
x_test_tfidf = tfidf.transform(x_test) # Convert testing data to vectors
```

### Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
tmp = [] # A list to maintain the mean of cross-value scores
cvs = [] # A list to maintain the cross-value scores

for i in range(1, 11): # Loop to find the best 'max_depth (levels in a DT)' value
    clf = DecisionTreeClassifier(max_depth = i, criterion = "entropy").fit(x_train_tfidf, y_train) # Compute DT
    cvs = cross_val_score(clf, x_train_tfidf, y_train, cv = 5) # Find cross-value score for max_depth=i
    tmp.append(cvs.mean())
```

```python
# Find the maximum mean cross-vaue score
for i in range(len(tmp)):
    if tmp[i] == np.max(tmp):
        print(i+1, ' : ', tmp[i])
```

```python
# As 9 is the max_depth value for which the cross-value score was maximum, using that compute DT
DT_clf = DecisionTreeClassifier(max_depth = 9, criterion = "entropy").fit(x_train_tfidf, y_train)
```

```python
y_pred = DT_clf.predict(x_test_tfidf) # Prediction on test data using the DT model
y_pred
```

```python
# Find accuracy score
from sklearn.metrics import accuracy_score
accuracy_score_DT = accuracy_score(y_test, y_pred)
accuracy_score_DT
```

```python
# Compute confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# Print the classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
tmp = [] # A list to maintain the mean of cross-value scores
cvs = [] # A list to maintain the cross-value scores

for i in range(100, 501, 100): # Loop to find the best number of decision trees (n_estimators) to combine
    clf = RandomForestClassifier(max_depth=10, criterion='entropy', n_estimators=i, random_state=2).fit(x_train_tfidf, y_train) # Compute RF
    cvs = cross_val_score(clf, x_train_tfidf, y_train, cv = 5) # Find cross-value score for n_estimators=i
    tmp.append(cvs.mean())
```

```python
# Find the maximum mean cross-vaue score
for i in range(len(tmp)):
    if tmp[i] == np.max(tmp):
        print(i+1, ' : ', tmp[i])
```

```python
# Mean cross-value score is maximum when n_estimators=300, so use that to fit the model
RF_clf = RandomForestClassifier(max_depth=10, criterion='entropy', n_estimators=300, random_state=2).fit(x_train_tfidf, y_train)
```

```python
y_pred = RF_clf.predict(x_test_tfidf) # Prediction on test data using the RF model
y_pred
```

```python
# Find accuracy score
from sklearn.metrics import accuracy_score
accuracy_score_RF = accuracy_score(y_test, y_pred)
accuracy_score_RF
```

```python
# Compute confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# Print classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

### Neural Networks

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score
```

```python
tmp = [] # Empty list which will be needed to store mean of cross-val scores
for i in range(1, 6): # Outer-loop runs for the number of hidden layers
    for j in range(1, 6): # Inner-loop runs for the number of hidden units in each layer
        clf = MLPClassifier(hidden_layer_sizes=(j,i), random_state=2, max_iter=500, alpha=0.01) # NNet CLassifier
        scores = cross_val_score(clf, x_train_tfidf, y_train, cv=5) # 5-fold Cross Validation
        print('Cross Validation Score for ', i, ' layer(s) and ', j, ' unit(s): ', scores)
        tmp.append(scores.mean()) # This list has the mean cross-val scores
```

```python
tmp
```

```python
np.max(tmp) # Get the maximum cross-val score
```

```python
# Find the index which has the maximum cross-val score
for i in range(25):
    #print(i, ' : ', tmp[i])
    if tmp[i] == 0.9943761051753288:
        print(i+1, ' : ', tmp[i])
```

```python
# 1 hidden layers with 3 hidden units is the ideal solution here.
```

```python
MLP_clf = MLPClassifier(hidden_layer_sizes=(3,1), random_state=2, max_iter=500, alpha=0.01).fit(x_train_tfidf, y_train)
```

```python
y_pred = MLP_clf.predict(x_test_tfidf) # Prediction on test data using MLP model
y_pred
```

```python
# Find accuracy score
from sklearn.metrics import accuracy_score
accuracy_score_MLP = accuracy_score(y_test, y_pred)
accuracy_score_MLP
```
